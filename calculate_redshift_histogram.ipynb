```python
file_path = "/home/jeremiasrodriguez/data/astroclip_desi.1.1.5.h5"
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Open the HDF5 file
all_redshifts = []

with h5py.File(file_path, 'r') as h5_file:
    # Loop through each group in the file and concatenate redshifts
    for group_name in h5_file.keys():
        redshift_data = h5_file[f"{group_name}/redshifts"][:]
        all_redshifts.append(redshift_data)

# Combine all redshifts into a single array
combined_redshifts = np.concatenate(all_redshifts)

# Plot the histogram of redshifts
plt.figure(figsize=(10, 6))
plt.hist(combined_redshifts, bins=200, color='blue', alpha=0.7)
plt.title('Histogram of Redshift')
plt.xlabel('Redshift')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print("This histogram comes from ", len(combined_redshifts), "samples")
```
